# Data Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Data/fifa_players_22_v2.csv")
df
```

```python
list(df.columns)
```

```python
df.isna().sum()
```

```python
# Data Cleaning
# Dropping unnecessary columns

col_to_drop = ['Unnamed: 0', 'sofifa_id', 'player_url', 'long_name', 'club_team_id', 'nation_team_id', 'real_face', 'ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 'rw', 'lam', 'cam', 'ram', 'lm', 'lcm', 'cm', 'rcm', 'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb', 'rcb', 'rb', 'gk', 'player_face_url', 'club_logo_url', 'club_flag_url', 'nation_logo_url', 'nation_flag_url']
```

```python
# Setting a cleaned up player dataframe
players22 = df.drop(col_to_drop, axis=1)
players22
```

```python
players22.info()
```

```python
col_to_drop2 = ['club_jersey_number', 'club_loaned_from', 'club_joined', 'club_contract_valid_until', 'nation_position', 'nation_jersey_number']
players22.drop(col_to_drop2, axis=1, inplace=True)
```

```python
players22.isna().sum()
```

```python
players22.rename(columns={'short_name':'name'}, inplace=True)
```

```python
# Fill in missing values with 'none' for columns ('player_tags', 'player_traits')
players22['player_tags'].fillna('NA', inplace=True)
players22['player_traits'].fillna('NA', inplace=True)
```

```python
# Drop missing values in columns ('value_eur', 'wage_eur', 'club_name', 'league_name', 'league_level', 'club_position')
players22.dropna(subset=['value_eur', 'wage_eur'], inplace=True)
```

```python
players22
```

```python
# Fill in missing values with '0' for columns ('release_clause_eur')
players22['release_clause_eur'].fillna(0, inplace=True)
```

```python
players22.info()
```

- All the players, except GK,  are missing the goalkeeping speed due to their positions.
